```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('000668_0.jpg',0)
```

```python
print(img.shape)
```

```python
cv2_imshow(img) 
```

```python
import numpy as np
sobel_2 = np.array([[-1,0,1], [-2,0,2],[-1,0,1]])
```

```python
edge_sobel2 = np.zeros(img.shape)
for i in range(1,img.shape[0]-1):
  for j in range(1,img.shape[1]-1):
    mat = np.multiply(sobel_2,img[i-1:i+2,j-1:j+2]);
    edge_sobel2[i,j] = (abs(np.sum(mat)));
 
```

```python
cv2_imshow(edge_sobel2) 
```
